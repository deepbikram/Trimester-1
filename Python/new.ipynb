```python
a = [2, 4, 6, 8, 10]

# add a single item to the end
a.append(12)
print(a)        # [2, 4, 6, 8, 10, 12]

# add many items at once to the end
a.extend([14, 16])
print(a)        # [2, 4, 6, 8, 10, 12, 14, 16]

# add (insert) an item at a specific position
a.insert(2, 5) 
print(a)        # [2, 4, 5, 6, 8, 10, 12, 14, 16]
```
